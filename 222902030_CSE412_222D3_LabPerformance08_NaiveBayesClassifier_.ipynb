```python
# 100 sample documents (50 sports + 50 politics)
documents = []

# 50 sports articles
sports_articles = [
    'The team won the championship.',
    'He scored a goal in the last minute.',
    'The cricket match was intense.',
    'The football team played well.',
    'The coach praised the players.',
    'They won the tennis match.',
    'The baseball game ended in a tie.',
    'She broke the swimming record.',
    'The basketball league is exciting.',
    'He trained hard for the marathon.',
    'The player received a red card.',
    'The striker scored twice.',
    'The athletes are preparing for the Olympics.',
    'The referee stopped the match.',
    'The race ended in a photo finish.',
    'The gymnast performed well.',
    'The tournament will start next week.',
    'They bought a new stadium.',
    'The sports fans are excited.',
    'The boxing match was brutal.',
    'The new player joined the club.',
    'The captain led the team to victory.',
    'She won the gold medal.',
    'The cycling event attracted many.',
    'The rugby match was thrilling.',
    'The players signed new contracts.',
    'The championship was held in London.',
    'The team practiced every day.',
    'The athletes were interviewed.',
    'The victory was celebrated.',
    'The volleyball final is tomorrow.',
    'He set a new running record.',
    'The crowd cheered loudly.',
    'The referee announced the winner.',
    'The players received awards.',
    'The training camp starts soon.',
    'The team wore new uniforms.',
    'The club bought new players.',
    'The tennis finals were exciting.',
    'The players scored quickly.',
    'The audience enjoyed the match.',
    'The team reached the finals.',
    'The stadium was full.',
    'The goalkeeper saved the goal.',
    'The player dribbled past defenders.',
    'The manager changed tactics.',
    'The tournament featured top teams.',
    'The match lasted for two hours.',
    'The sports community reacted positively.',
    'The cricket fans watched live.'
]

# 50 politics articles
politics_articles = [
    'The election results were announced.',
    'The government passed a new law.',
    'The president gave a speech.',
    'The parliament debated the issue.',
    'The minister resigned yesterday.',
    'The political party won the election.',
    'The prime minister visited the city.',
    'The leaders signed the agreement.',
    'The new policy was implemented.',
    'The cabinet discussed the reforms.',
    'The lawmakers voted unanimously.',
    'The president addressed the nation.',
    'The campaign focused on education.',
    'The government plans to reduce taxes.',
    'The opposition criticized the proposal.',
    'The mayor announced new projects.',
    'The leaders held a press conference.',
    'The law was passed by the parliament.',
    'The election campaign started today.',
    'The candidate promised healthcare reforms.',
    'The prime minister met foreign leaders.',
    'The new bill was introduced.',
    'The ruling party celebrated victory.',
    'The budget was presented in parliament.',
    'The politicians debated healthcare policies.',
    'The president signed the bill.',
    'The opposition leader spoke to media.',
    'The minister announced education plans.',
    'The government changed immigration laws.',
    'The politicians held a debate.',
    'The prime minister spoke to reporters.',
    'The bill faced opposition.',
    'The campaign was widely discussed.',
    'The government focused on economic growth.',
    'The cabinet approved the new project.',
    'The leaders discussed international relations.',
    'The new tax law took effect.',
    'The parliament passed the budget.',
    'The politicians proposed amendments.',
    'The prime minister introduced reforms.',
    'The opposition staged a protest.',
    'The government held public consultations.',
    'The president inaugurated the event.',
    'The cabinet members met yesterday.',
    'The foreign minister visited abroad.',
    'The leaders met for summit talks.',
    'The mayor launched new services.',
    'The prime minister held press talks.',
    'The politicians focused on development.',
    'The government announced welfare programs.'
]

# Merge sports and politics articles
documents = sports_articles + politics_articles

# Show original documents
print(documents)
print("\n")

# Create labels: 0 = sports, 1 = politics
labels = [0]*50 + [1]*50

# Convert documents to lowercase
lower_case_documents = [doc.lower() for doc in documents]

# Show lowercase documents
print("All Document print is lowercase letter: \n")
print(lower_case_documents)
print("\n")

# Remove punctuation from documents
import string
punctuation_removed_documents = [
    "".join(char for char in doc if char not in string.punctuation)
    for doc in lower_case_documents
]

# Show documents without punctuation
print("After Remove Punctuation word in the list: \n")
print(punctuation_removed_documents)
print("\n")

# Split documents into words
preprocessed_documents = [doc.split(' ') for doc in punctuation_removed_documents]

# Show tokenized documents
print("After Split into words: \n")
print(preprocessed_documents)
print("\n")

# Count each word in documents
from collections import Counter
frequency_list = [Counter(doc) for doc in preprocessed_documents]

# Show word count per document
print("Count the every word in the list: \n")
print(frequency_list)
print("\n")

# Vectorize documents (convert text to numbers)
from sklearn.feature_extraction.text import CountVectorizer
count_vector = CountVectorizer()
count_vector.fit(documents)

# Show unique words (features)
print(count_vector.get_feature_names_out())
print("\n")

# Transform documents into document-term matrix
doc_array = count_vector.transform(documents).toarray()

# Show document-term matrix
print("Convert in array format: \n")
print(doc_array)
print("\n\n")

# Display matrix as table
import pandas as pd
frequency_matrix = pd.DataFrame(doc_array, columns=count_vector.get_feature_names_out())
print(frequency_matrix)

# Import libraries for model training
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix
import numpy as np

# Split data (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(documents, labels, random_state=1)

# Vectorize training and test data
count_vector = CountVectorizer()
train = count_vector.fit_transform(X_train)
test = count_vector.transform(X_test)

# Train Naive Bayes model
naive_bayes = MultinomialNB()
naive_bayes.fit(train, y_train)

# Predict test results
predictions = naive_bayes.predict(test)

# Calculate accuracy
accuracy = np.mean(predictions == y_test)

# Show accuracy
print("Accuracy : \n")
print(accuracy)
print("\n")

# Show confusion matrix
print("Confusion Matric: \n")
print(confusion_matrix(y_test, predictions))
```
