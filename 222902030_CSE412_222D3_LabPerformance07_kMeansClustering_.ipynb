```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load dataset from GitHub
df = pd.read_csv("https://raw.githubusercontent.com/rakesh22230/Machine_Learning_Practise/refs/heads/main/Mall_Customers.csv")

print("First 5 rows of dataset:")
print(df.head())  # Show first few rows to understand data
print("\n")

# Choose features for clustering
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]

# Visualize the data (before clustering)
plt.figure(figsize=(8, 5))
plt.scatter(X['Annual Income (k$)'], X['Spending Score (1-100)'], c='blue')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Mall Customer Data Before Clustering')
plt.grid(True)
plt.show()

print("\n")

# Apply K-Means Clustering with 5 clusters
kmeans_model = KMeans(n_clusters=5, init='k-means++', random_state=42)
kmeans_model.fit(X)

# Predict cluster labels
predictedY = kmeans_model.predict(X)

# Print predicted cluster labels for each customer
print("Predicted cluster labels for each customer:")
print(predictedY)
print("\n")

# Optionally, add cluster labels back to dataframe
df['Cluster'] = predictedY

# Plot before and after clustering side-by-side
plt.figure(figsize=(12, 5))

# Before clustering (all points gray)
plt.subplot(1, 2, 1)
plt.scatter(
    X['Annual Income (k$)'],
    X['Spending Score (1-100)'],
    color='gray',
    s=50
)
plt.title('Before Clustering (Mall Customer)')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')

# After clustering (colored by cluster)
plt.subplot(1, 2, 2)
plt.scatter(
    X['Annual Income (k$)'],
    X['Spending Score (1-100)'],
    c=predictedY,  # Use predicted labels for color
    cmap='rainbow',
    s=50
)
plt.title('After Clustering (Mall Customer)')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')

plt.tight_layout()
plt.show()
```
