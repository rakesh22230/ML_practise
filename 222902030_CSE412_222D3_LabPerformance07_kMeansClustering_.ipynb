```python
import pandas as pd  # To load and work with data
import matplotlib.pyplot as plt  # To make plots
from sklearn.cluster import KMeans  # To do clustering

# Step 1: Load the data from GitHub
df = pd.read_csv("https://raw.githubusercontent.com/rakesh22230/Machine_Learning_Practise/refs/heads/main/Mall_Customers.csv")

# Step 2: Look at the first few rows to understand the data
print(df.head())

# Step 3: Choose the two columns we want to use for clustering
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]

# Step 4: Create the KMeans model to find 5 groups (clusters)
kmeans = KMeans(n_clusters=5, random_state=42)

# Step 5: Fit the model to the data (this finds the clusters)
kmeans.fit(X)

# Step 6: Get the cluster number for each customer
labels = kmeans.labels_

# Step 7: Make a plot to show the clusters
plt.figure(figsize=(8, 6))

# Scatter plot with color for each cluster
plt.scatter(
    X['Annual Income (k$)'],
    X['Spending Score (1-100)'],
    c=labels,  # Color points by cluster label
    cmap='rainbow',  # Use a colorful palette
    s=50  # Size of points
)

# Step 8: Add title and labels
plt.title('Customer Clusters')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')

# Step 9: Show the plot
plt.show()
```
