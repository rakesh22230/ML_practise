```python
# List of sample documents to be processed
documents = [
    'The canal is open to shipping.',
    'This is a ship shipping ship, shipping shipping ships.',
    'The ship locked into the new canal.'
]

# Print the original list of documents
print(documents)
print("\n")

# Convert all documents to lowercase to maintain uniformity in processing
lower_case_documents = [doc.lower() for doc in documents]

# Output the documents after converting to lowercase
print("All Document print is lowercase letter: \n")
print(lower_case_documents)
print("\n")

import string
# Remove punctuation from each document to standardize the words for analysis
punctuation_removed_documents = [
    "".join(char for char in doc if char not in string.punctuation)
    for doc in lower_case_documents
]

# Display the documents after removing punctuation
print("After Remove Punctuation word in the list: \n")
print(punctuation_removed_documents)
print("\n")

# Tokenize each document by splitting them into individual words
preprocessed_documents = [doc.split(' ') for doc in punctuation_removed_documents]

# Output the documents after splitting into words (tokenized form)
print("After Split into words: \n")
print(preprocessed_documents)
print("\n")

# Count the frequency of each word in the documents using Counter
from collections import Counter
frequency_list = [Counter(doc) for doc in preprocessed_documents]

# Display the frequency of each word in the documents
print("Count the every word in the list: \n")
print(frequency_list)
print("\n")

# Initialize CountVectorizer for extracting word counts (features)
from sklearn.feature_extraction.text import CountVectorizer
count_vector = CountVectorizer()
count_vector.fit(documents)

# Display the list of unique features (words) found in the documents
print(count_vector.get_feature_names_out())
print("\n")

# Transform the documents into an array of word counts (document-term matrix)
doc_array = count_vector.transform(documents).toarray()

# Display the document-term matrix (array format)
print("Convert in array format: \n")
print(doc_array)
print("\n\n")

# Import pandas to display the frequency matrix as a DataFrame
import pandas as pd
frequency_matrix = pd.DataFrame(doc_array, columns=count_vector.get_feature_names_out())

# Display the frequency matrix as a DataFrame
print(frequency_matrix)

# Import necessary libraries for working with a real dataset
import pandas as pd

# Load the SMS Spam dataset from the URL
df = pd.read_table('https://raw.githubusercontent.com/rakesh22230/Machine_Learning_Practise/refs/heads/main/SMSSpamCollection', sep='\t', header=None, names=['label', 'message'])

# Display the first few rows of the dataset
df.head()

print("\n\n")
# Map the 'ham' and 'spam' labels to 0 and 1 respectively for classification
df['label'] = df['label'].map({'ham': 0, 'spam': 1})

# Display the dataset after label transformation
print(df.head())
print("\n")

# Import libraries for machine learning model training and evaluation
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix
import numpy as np

# Split the dataset into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(df['message'], df['label'], random_state=1)

# Convert text data into numerical feature vectors using CountVectorizer
count_vector = CountVectorizer()
train = count_vector.fit_transform(X_train)
test = count_vector.transform(X_test)

# Initialize the Naive Bayes classifier and train it with the training data
naive_bayes = MultinomialNB()
naive_bayes.fit(train, y_train)

# Make predictions on the test data using the trained model
predictions = naive_bayes.predict(test)

# Calculate the accuracy of the model by comparing predictions with actual labels
accuracy = np.mean(predictions == y_test)

# Output the accuracy of the Naive Bayes model
print("Accuracy : \n")
print(accuracy)
print("\n")

# Print confusion matrix to evaluate the performance of the classification model
print("Confusion Matric: \n")
print(confusion_matrix(y_test, predictions))
```
